**Step 4: Loading the Dataset**

In this step, I loaded the Titanic dataset into the Python environment using the Pandas library.
Loading the dataset allows us to view and analyze the data.
The dataset file used in this project is named **titanic.csv**.

After loading the dataset, the first five rows of the dataset were displayed to understand the structure of the data.


```python
import pandas as pd

df = pd.read_csv("titanic.csv")

df.head()
```

**Step 5: Checking Dataset Information**

In this step, I examined the structure and summary of the dataset.
The purpose of this step is to understand the number of rows and columns, data types of each column, and identify missing values in the dataset.

The `info()` function provides details about the dataset structure, while the `describe()` function gives statistical summaries such as mean, minimum, maximum, and standard deviation.


```python
df.info()

df.describe()
```

**Step 6: Checking Missing Values**

In this step, I checked for missing values in the dataset.
Missing values can affect the accuracy of data analysis, so it is important to identify them before performing further analysis.

The `isnull().sum()` function was used to count the number of missing values in each column of the dataset.


```python
df.isnull().sum()
```

**Step 7: Handling Missing Values in the Age Column**

In this step, I handled the missing values in the Age column.
Missing values in the Age column were replaced using the mean (average) value of the Age column.

Using the mean value helps maintain the overall distribution of the data and allows further analysis to be performed without errors.


```python
df["Age"].fillna(df["Age"].mean(), inplace=True)

df["Age"].isnull().sum()
```

**Step 8: Dropping the Cabin Column**

In this step, I removed the Cabin column from the dataset.
The Cabin column contained a large number of missing values, making it difficult to use for analysis.

Dropping this column helps simplify the dataset and improves the quality of data analysis.


```python
df.drop("Cabin", axis=1, inplace=True)

df.columns
```

**Step 9: Handling Missing Values in the Embarked Column**

In this step, I handled the missing values in the Embarked column.
The missing values were replaced using the mode (most frequent value) of the column.

Using the mode is appropriate for categorical data because it represents the most common category in the dataset.


```python
df["Embarked"].fillna(df["Embarked"].mode()[0], inplace=True)

df["Embarked"].isnull().sum()
```

**Step 10: Analyzing Survival by Gender**

In this step, I analyzed the survival of passengers based on gender.
The dataset was grouped by the Sex column, and the total number of survivors was calculated for each gender.

This analysis helps us understand whether gender had an impact on survival during the Titanic disaster.


```python
df.groupby("Sex")["Survived"].sum()
```

**Step 10A: Verifying Gender Distribution**

In this step, I verified the number of male and female passengers in the dataset.
This check was performed to ensure that the dataset contains both genders and to validate the accuracy of the previous analysis results.


```python
df["Sex"].value_counts()
```

**Step 11: Visualizing Survival by Gender**

In this step, I created a bar chart to visualize the survival of passengers based on gender.
The bar chart helps to clearly compare the number of survivors among male and female passengers.

Data visualization makes it easier to understand patterns and trends in the dataset.


```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x="Sex", y="Survived", data=df)

plt.title("Survival by Gender")

plt.show()
```

**Step 12: Analyzing Survival by Passenger Class**

In this step, I analyzed the survival of passengers based on their passenger class (Pclass).
The dataset was grouped by passenger class, and the average survival rate was calculated for each class.

This analysis helps determine whether the passenger class had an impact on survival chances during the Titanic disaster.


```python
df.groupby("Pclass")["Survived"].mean()
```

**Step 13: Visualizing Survival by Passenger Class**

In this step, I created a bar chart to visualize the survival rate of passengers based on their passenger class (Pclass).
The bar chart helps compare survival chances among first, second, and third-class passengers.

This visualization makes it easier to understand how passenger class influenced survival during the Titanic disaster.


```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x="Pclass", y="Survived", data=df)

plt.title("Survival by Passenger Class")

plt.show()
```

**Step 14: Visualizing Age Distribution of Passengers**

In this step, I created a histogram to visualize the distribution of passenger ages.
The histogram shows how passengers are distributed across different age groups.

This visualization helps identify patterns such as the most common age range among passengers.


```python
import matplotlib.pyplot as plt

plt.hist(df["Age"])

plt.title("Age Distribution of Passengers")

plt.xlabel("Age")

plt.ylabel("Number of Passengers")

plt.show()
```
