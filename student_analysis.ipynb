# Data analysis - Task 1
## Student Performance Data Analysis
In this project, we are using the "Student Performance Dataset" to analyze how various factors like study time, gender, and absences impact the final grades (G3) of students.


### Step 1: Importing Necessary Libraries
We start by importing the essential Python libraries:
* **Pandas**: For data manipulation and analysis.
* **Matplotlib & Seaborn**: For creating data visualizations (graphs).

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
!unzip student.zip
```

### Step 2: Loading the Dataset
We are loading the `student-mat.csv` file. Since the data fields in this file are separated by semicolons, we use the `sep=';'` parameter.

```python
df = pd.read_csv('student-mat.csv', sep=';')
df.head()
```

### Step 3: Data Inspection and Cleaning
In this step, we check the structure of the data and look for any missing (null) values to ensure the data is clean before analysis.

```python
df.info()
print("\nMissing values check:")
print(df.isnull().sum())
```

### Step 4: Statistical Summary
We generate a mathematical summary (Mean, Max, Min, etc.) of the numeric columns to understand the general range of student ages and grades.

```python
# Numbers wale columns ka summary dekhne ke liye
df.describe()
```

### Step 5: Distribution of Final Grades (G3)
This histogram shows the frequency of final marks. It helps us see the overall performance of the class and identify how many students scored high or low.

```python
# Final Grade (G3) ka distribution plot
plt.figure(figsize=(8, 5))
sns.histplot(df['G3'], kde=True, color='blue')
plt.title('Distribution of Final Grades (G3)')
plt.xlabel('Final Grade (G3)')
plt.ylabel('Number of Students')
plt.show()
```

### Step 6: Comparison of Grades by Gender
We use a boxplot to compare the performance of Male (M) and Female (F) students to see if there is a significant difference in their average scores.

```python
# Sex (Gender) vs Final Grade (G3)  box plot
plt.figure(figsize=(8, 6))
sns.boxplot(x='sex', y='G3', data=df, palette='Set2')
plt.title('Comparison of Final Grades by Gender')
plt.xlabel('Gender (F: Female, M: Male)')
plt.ylabel('Final Grade (G3)')
plt.show()
```

### Step 7: Impact of Study Time on Final Grades
This bar plot visualizes the relationship between the time spent studying and the final grades achieved. It helps confirm if more study hours lead to better results.

```python
# Study time vs Final Grade bar plot
plt.figure(figsize=(8, 5))
sns.barplot(x='studytime', y='G3', data=df, palette='viridis')
plt.title('Effect of Study Time on Final Grades')
plt.xlabel('Study Time (1: <2 hrs, 2: 2-5 hrs, 3: 5-10 hrs, 4: >10 hrs)')
plt.ylabel('Average Final Grade (G3)')
plt.show()
```

### Step 8: Correlation Heatmap
This is a crucial step that shows the correlation between all numeric variables. It helps us identify which factors (like previous grades or absences) have the strongest impact on the final grade.

```python
# only select the numbers columns
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Make Correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Student Data')
plt.show()
```

## Final Conclusion
1. **Previous Grades**: Grades G1 and G2 show the strongest positive correlation with the final grade (G3).
2. **Study Habit**: Students who study more than 5 hours a week tend to have higher average scores.
3. **Absences**: There is a negative trend indicating that higher absences often lead to lower final grades.
